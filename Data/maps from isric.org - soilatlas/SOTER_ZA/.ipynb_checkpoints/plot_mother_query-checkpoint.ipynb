# Reduce Data, Drop Duplicates & Rename Columns

```python
# reduce dataframes
reduced_geo_poly_data = geo_poly_data[['SUID','LITHOLOGY','LANDFORM','SOIL','geometry']]
reduced_query_data = query_data[['SOTER_ID','RELIEF_I','BEDROCK_DEPTH']]
```

```python
# drop duplicates from reduced_geo_poly_data in SUID column
reduced_geo_poly_data = reduced_geo_poly_data.drop_duplicates(subset=['SUID'])
reduced_geo_poly_data.head(3)
```

```python
# recheck for duplicates in reduced_geo_poly_data
print('total: ', reduced_geo_poly_data.shape[0])
print('unique: ', len(reduced_geo_poly_data.SUID.unique()))
```

```python
# rename
reduced_geo_poly_data = reduced_geo_poly_data.rename(columns={'SUID':'SOTER_ID'})
reduced_geo_poly_data = reduced_geo_poly_data.set_index('SOTER_ID')
reduced_query_data = reduced_query_data.set_index('SOTER_ID')
```

# Shuffle Columns

```python
final_data.shape
```

# Convert to GeoPandas Dataframe And Plot

```python
geo_dataframe = gpd.GeoDataFrame(final_data)
geo_dataframe.head(5)
```

```python
# basic plot - plot the data using geopandas .plot() method
fig, ax = plt.subplots(figsize = (15,15))
geo_dataframe.plot(ax=ax,
           linewidth = 0.3,
           edgecolor='black',
           facecolor='lightgrey',
           alpha=0.9)
plt.show()
```

# Get Data From Input Coordinates

```python
# find info at coordinates

#input coordinate
x = [29,25,17.5,30]
y = [-27,-32,-30,-24.1]
names = ['1','2','3','4']

df_dict = {'Name': names, 'x': x, 'y': y}
df_dict

df = pd.DataFrame.from_dict(df_dict)
inp_coords = gpd.GeoDataFrame(df, geometry=gpd.points_from_xy(df.x, df.y))
inp_coords
```

```python
# reset index
geo_dataframe = geo_dataframe.reset_index()
```

```python
# search for point in all polygons
return_polys_ind = []
answer_polys = gpd.GeoDataFrame()


for j in range(inp_coords.shape[0]):

    coord_geo = inp_coords.iloc[j].geometry

    for i in range(geo_dataframe.shape[0]):
        if coord_geo.within(geo_dataframe.iloc[i].geometry):
            return_polys_ind.append(i)

    answer_polys = answer_polys.append(geo_dataframe.iloc[return_polys_ind[j]])
answer_polys
```

```python
# drop duplicates from answer_polys
answer_polys = answer_polys.drop_duplicates(subset=['geometry'],keep='last')
answer_polys
```

```python
out_dict = {'Name': inp_coords.Name.values,
            'x': inp_coords.x.values,
            'y': inp_coords.y.values,
            'BEDROCK_DEPTH': answer_polys.BEDROCK_DEPTH.values,
            'LANDFORM': answer_polys.LANDFORM.values,
            'LITHOLOGY': answer_polys.LITHOLOGY.values,
            'SOIL' : answer_polys.SOIL.values,
            #'MAX_EL': answer_polys.MAX_EL.values,
            #'MIN_EL': answer_polys.MIN_EL.values,
            'RELIEF_I': answer_polys.RELIEF_I.values}#,
            #'VEGETATION' : answer_polys.VEGETATION.values,
            #'PROPORTION': answer_polys.PROPORTION.values}

final_out = pd.DataFrame(out_dict)
final_out
```

# Final Plot

```python
# plot within SA
reduced_set = geo_dataframe.SOTER_ID.unique()
cols = ['b', 'g', 'r', 'c', 'm', 'y', 'pink', 'lightblue', 'pink', 'orange', 'lightgreen', 'darkgreen', 'lightgrey', 'darkblue', 'purple']
cols = cols*10

fig, ax = plt.subplots(figsize = (15,15))
leg_handles = []
transparency = 0.9

# plot all geometry in geo_dataframe for background
geo_dataframe.plot(ax=ax,
           linewidth = 0.3,
           edgecolor='black',
           facecolor='lightgrey',
           alpha=transparency)

# plot found areas
for i,geo in enumerate(answer_polys.geometry.unique()):
    curr_poly = answer_polys[answer_polys.geometry == geo]
    curr_poly.plot(ax=ax,
                   linewidth = 0.5,
                   edgecolor='black',
                   alpha=0.3,
                   facecolor=cols[i])#,
                   #hatch='//')
    leg_handles.append(mpatches.Patch(color=cols[i], label=final_out.Name[i], alpha=0.3))

# plot coordinates
for i,coor in enumerate(inp_coords.geometry.unique()):
    curr_coor = inp_coords[inp_coords.geometry == coor]
    curr_coor.plot(ax=ax, color=cols[i], linewidth = 1, edgecolor='black', marker='o', markersize=80)
    leg_handles.append(mlines.Line2D([], [],color=cols[i], linewidth = 1, marker='o',linestyle='None', label=inp_coords.Name[i]))

# other details
plt.legend(handles=leg_handles, loc='upper left')
ax.set_title('Coordinates Summary - SOTER Data', fontsize=20)
ax.grid(color='grey',linestyle='-', linewidth=0.5, alpha=0.4)
#ax.set_axis_off()    # turn off the x and y axis
#plt.axis('equal')    # ensure the x and y axis are uniformly spaced
plt.show()
```

# Zoomed Plot

```python
# zoom on point p
p = 3
scale_factor = 0.2
x_min = inp_coords.iloc[p].x - scale_factor
x_max = inp_coords.iloc[p].x + scale_factor
y_min = inp_coords.iloc[p].y - scale_factor
y_max = inp_coords.iloc[p].y + scale_factor

# plot within SA
reduced_set = geo_dataframe.SOTER_ID.unique()
cols = ['b', 'g', 'r', 'c', 'm', 'y', 'pink', 'lightblue', 'pink', 'orange', 'lightgreen', 'darkgreen', 'lightgrey', 'darkblue', 'purple']
cols = cols*10

fig, ax = plt.subplots(figsize = (15,15))
leg_handles = []
transparency = 0.9

# plot all geometry in geo_dataframe for background
geo_dataframe.plot(ax=ax,
           linewidth = 0.3,
           edgecolor='black',
           facecolor='lightgrey',
           alpha=transparency)

# plot found areas
for i,geo in enumerate(answer_polys.geometry.unique()):
    curr_poly = answer_polys[answer_polys.geometry == geo]
    curr_poly.plot(ax=ax,
                   linewidth = 0.5,
                   edgecolor='black',
                   alpha=0.3,
                   facecolor=cols[i])#,
                   #hatch='//')
    leg_handles.append(mpatches.Patch(color=cols[i], label=final_out.Name[i], alpha=0.3))

# plot coordinates
for i,coor in enumerate(inp_coords.geometry.unique()):
    curr_coor = inp_coords[inp_coords.geometry == coor]
    curr_coor.plot(ax=ax, color=cols[i], linewidth = 1, edgecolor='black', marker='o', markersize=80)
    leg_handles.append(mlines.Line2D([], [],color=cols[i], linewidth = 1, marker='o',linestyle='None', label=inp_coords.Name[i]))

# other details
plt.legend(handles=leg_handles, loc='upper left')
ax.set_title('Coordinates Summary - SOTER Data', fontsize=20)
ax.grid(color='grey',linestyle='-', linewidth=0.5, alpha=0.4)
#ax.set_axis_off()    # turn off the x and y axis
#plt.axis('equal')    # ensure the x and y axis are uniformly spaced
ax.set_xlim(x_min, x_max)
ax.set_ylim(y_min, y_max)
plt.show()
```
